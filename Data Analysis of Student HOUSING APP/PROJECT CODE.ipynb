13.**What is the total monthly rent of your unit?**:
1. Detected 31 missing fields, addressing them with the median value.
2. Implemented a series of data cleaning operations on this column:

    - **Column Identification**: Named the column for easier reference.
    - **Missing Value Check**: Verified for missing entries before initiating the cleaning.
    - **Numeric Conversion**: Transformed all column entries to numeric, turning non-numeric entries to NaN.
    - **Median Calculation**: Computed the median rent, post-conversion to numeric, to only include valid numbers.
    - **Filling Missing Values**: Populated missing entries with the calculated median rent.
    - **Removal of Non-numeric Characters**: Eradicated non-numeric symbols like dollar signs and commas, converting entries to float.
    - **Invalid Rent Values**: Marked "4 months" entries as NaN for being invalid rent amounts.
    - **Capping Rent Values**: Assigned NaN to rents over $3000, considering them unrealistic for the dataset.
    - **Median Recalculation**: Recalculated the median rent after excluding unrealistic figures and updated NaNs with this revised median.
    - **Descriptive Statistics**: Reviewed the column's descriptive stats to evaluate the cleaning outcome.

```python
# Define the column name
rent_column_name = "What is the total monthly rent of your unit?"

# Check Missing Values Before Cleaning
missing_values_before = questionnaire_responses[rent_column_name].isnull().sum()
print(f"Missing values in '{rent_column_name}' before cleaning:", missing_values_before)

# Ensure that all non-numeric data is set to NaN
questionnaire_responses[rent_column_name] = pd.to_numeric(questionnaire_responses[rent_column_name], errors='coerce')

# Now calculate the median of the column after cleaning
median_rent = questionnaire_responses[rent_column_name].median()

# Fill in the missing values with the median
questionnaire_responses[rent_column_name].fillna(median_rent, inplace=True)

# Check Missing Values After Filling
missing_values_after = questionnaire_responses[rent_column_name].isnull().sum()
print(f"Missing values in '{rent_column_name}' after filling:", missing_values_after)

# Remove non-numeric characters and convert to float
questionnaire_responses['What is the total monthly rent of your unit?'] = (
    questionnaire_responses['What is the total monthly rent of your unit?']
    .replace('[\$,]', '', regex=True)  # Remove dollar signs and commas
    .replace('4 months', np.nan, regex=False)  # Replace '4 months' with NaN
    .replace(r'^\s*$', np.nan, regex=True)  # Replace empty strings with NaN
    .astype(float)  # Convert to float
)

# Replace values that are unrealistically high with NaN
questionnaire_responses.loc[
    questionnaire_responses['What is the total monthly rent of your unit?'] > 3000,
    'What is the total monthly rent of your unit?'
] = np.nan

# Replace NaN values with the median rent
median_rent = questionnaire_responses['What is the total monthly rent of your unit?'].median()
questionnaire_responses['What is the total monthly rent of your unit?'].fillna(median_rent, inplace=True)

# Now, you can check the results
print(questionnaire_responses['What is the total monthly rent of your unit?'].describe())
```

16.**If yes, what is your monthly rental insurance?**:
#### Identifying and Replacing Values
- **Zero Replacement**: Missing and non-numeric values in the monthly rental insurance column are replaced with zero, assuming these indicate no payment.
- **Range Adjustment**: Values outside the practical range of 0 to $5,000 are set to zero, aligning with the assumption of unreasonable insurance costs.

#### Analysis of Data Variability
- **Data Diversity**: The column exhibits a mix of numeric values, numeric strings (e.g., '13 dollars'), phrases suggesting no cost or unknown costs (e.g., 'No', 'Donâ€™t know exactly', 'na'), and illogical entries (like excessively long numbers).

#### Numeric Conversion and Value Adjustment
- **Conversion to Numeric**: Strings representing dollar amounts are converted to numeric values.
- **Absence of Cost Indicators**: Phrases that indicate no cost or unknown costs, such as 'No' and 'na', are considered as zero.
- **Upper Limit Enforcement**: A reasonable upper limit for insurance costs is established, with values exceeding this limit adjusted to NaN or 0, based on the context.
- **Imputation of Missing Values**: Missing values are imputed with a suitable statistic, like the median, derived from valid, non-erroneous entries.

```python
# Define the column name for monthly rental insurance
rental_insurance_column_name = "If yes, what is your monthly rental insurance?"

# Function to clean and convert insurance values
def clean_insurance_value(value):
    if pd.isna(value):
        return np.nan
    if isinstance(value, str):
        value = value.lower()
        if any(non_cost_phrase in value for non_cost_phrase in ['no', 'donâ€™t know', 'dont know', 'na', 'zero', '-', ' ']):
            return 0.0
        # Remove non-numeric characters and convert to float
        value = ''.join(filter(str.isdigit, value))
        return float(value) if value else 0.0
    return float(value)

# Apply the function to the rental insurance column
questionnaire_responses[rental_insurance_column_name] = questionnaire_responses[rental_insurance_column_name].apply(clean_insurance_value)

# Set an upper limit for insurance costs
upper_limit_insurance = 5000  # Set this based on what you deem to be a reasonable upper limit
questionnaire_responses.loc[questionnaire_responses[rental_insurance_column_name] > upper_limit_insurance, rental_insurance_column_name] = np.nan

# Impute missing values with median after cleaning
median_insurance = questionnaire_responses[rental_insurance_column_name].median()
questionnaire_responses[rental_insurance_column_name].fillna(median_insurance, inplace=True)

# Check missing values after cleaning and imputation
missing_values_after_cleaning = questionnaire_responses[rental_insurance_column_name].isnull().sum()
print(f"Missing values after cleaning and imputation in '{rental_insurance_column_name}':", missing_values_after_cleaning)
```
